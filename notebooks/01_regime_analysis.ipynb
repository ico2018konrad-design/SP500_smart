# Regime Analysis

This notebook shows regime scores over history.

```python
import sys
sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from src.data.yahoo_loader import load_spy, load_vix
from src.regime.detector import RegimeDetector

spy = load_spy('2015-01-01')
vix = load_vix('2015-01-01')
print(f'Loaded {len(spy)} SPY days, {len(vix)} VIX days')
```

```python
# Calculate regime score for each day
detector = RegimeDetector()
scores = []
dates = []
for i in range(200, len(spy)):
    spy_slice = spy.iloc[:i+1]
    vix_slice = vix.iloc[:i+1]
    try:
        result = detector.detect(spy_slice, vix_slice)
        scores.append(result.score)
        dates.append(spy.index[i])
    except Exception:
        scores.append(None)
        dates.append(spy.index[i])

regime_series = pd.Series(scores, index=dates)
print('Done. Regime scores calculated.')
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True)
ax1.plot(spy['Close']['2015':], label='SPY')
ax1.set_title('SPY Price')
ax1.legend()
ax2.plot(regime_series, label='Regime Score', color='purple')
ax2.axhline(y=8, color='green', linestyle='--', label='BULL threshold')
ax2.axhline(y=4, color='red', linestyle='--', label='BEAR threshold')
ax2.set_title('Regime Score (0-11)')
ax2.legend()
plt.tight_layout()
plt.savefig('../results/regime_analysis.png', dpi=100)
plt.show()
```
